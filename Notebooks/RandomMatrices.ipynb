```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def random_symmetric(n):
    base_matrix = np.random.normal(size=(n,n))
    transposed_matrix = np.transpose(base_matrix)
    return np.add(base_matrix,transposed_matrix)
```

```python
example_matrix = random_symmetric(3)
```

```python
np.linalg.eig(example_matrix)
```

```python
def random_matrix_experiment(n_trials=1000,n=3):
    eig_list = []
    for trial in range(n_trials):
        this_matrix = random_symmetric(n)
        
        eigen_values = list(np.linalg.eig(this_matrix)[0][0:n])
        
        eig_list = eig_list + eigen_values
    return eig_list
```

```python
example_eigenvalues = random_matrix_experiment(1000)
```

```python
plt.hist(random_matrix_experiment(1000,500),500)
```
