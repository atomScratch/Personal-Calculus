```python
import matplotlib.pyplot as plt
import random

def random_walk(n_steps=1000, deltaX = 1, deltaY = 1):
    step_list = [(0,0)]
    up_down_vals = [-1*deltaY,deltaY]
    for n in range(1,n_steps):
        step_list.append((deltaX*n, (step_list[n-1][1] + random.choice(up_down_vals))))
    return step_list
```
