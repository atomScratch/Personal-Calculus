```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.normal(size=(2,2))
```

```python
def run_random_matrix_experiment(n_trials = 1000, N=10):
    eig_list = []
    
    for trial in range(n_trials):
        active_matrix = np.random.normal(size=(N,N))
        eig_list.append(np.linalg.eig(active_matrix))
    
    return eig_list
```

```python
run_random_matrix_experiment()
```

```python
eig_list = run_random_matrix_experiment(100,1000)

reals = [np.real(z[1][1]) for z in eig_list]
imags = [np.imag(z[1][1]) for z in eig_list]

plt.scatter(reals, imags)
```

```python
eig_list = run_random_matrix_experiment(100,100)

reals = [np.real(z[1][1]) for z in eig_list]
imags = [np.imag(z[1][1]) for z in eig_list]

plt.scatter(reals, imags)
```

```python
eig_list = run_random_matrix_experiment(100,10)

reals = [np.real(z[1][1]) for z in eig_list]
imags = [np.imag(z[1][1]) for z in eig_list]

plt.scatter(reals, imags)
```

Notice that as N 9the height and width of the matrix) decreases, there is also some kind of inverse relation between N and the radius of the "circle."
