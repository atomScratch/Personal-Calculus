```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eulers(x_dot, v_dot, x_init, v_init, t_init = 0, step_size = 0.01,step_count=10000):
    x_vals = [x_init]
    v_vals = [v_init]
    t = t_init
    for step in range(step_count):
        active_slope_x = x_dot(x_vals[-1], v_vals[-1], t)
        x_vals.append(x_vals[-1] + active_slope_x *step_size)
        
        active_slope_v = v_dot(x_vals[-1], v_vals[-1], t)
        v_vals.append(v_vals[-1] + active_slope_v * step_size)
        
        t += step_size
    return x_vals, v_vals
```

```python
def xdot(x,v,t):
    return v

def vdot(x, v, t, F=1, freq=2, resonant_freq=1, gamma = 0.1):
    return F * np.cos(freq * t) - gamma * v - resonant_freq**2 * x
```

```python
xt_vals = eulers(xdot,vdot,1,1)[0]
```

```python
plt.plot(range(len(xt_vals)), xt_vals)
```
