```python
import numpy as np
import matplotlib.pyplot as plt

def convolve(f, g, support = 100):
    partition = np.linspace(-support,support,support*100)
    change_x = partition[1] - partition[0]
    def h(z):
        area_sum = 0
        for x in range(len(partition)):
            area_sum += f(partition[x])*g(z-partition[x])*change_x
        return area_sum
    return h
```

```python
def uniform(x):
    if 0<=x<=1:
        return 1
    else:
        return 0
    
h2 = convolve(uniform,uniform,support=1)

x = np.linspace(0, 2, 100)
y = np.array([h2(i) for i in x])

plt.plot(x,y)
```

```python
h3 = convolve(h2,uniform,support=2)

x = np.linspace(0, 3, 100)
y = np.array([h3(i) for i in x])

plt.plot(x,y)
```

```python
def exponential(x,lam=1):
    if x<=0:
        return 0
    else:
        return lam * np.e ** (-lam * x)

h4 = convolve(exponential,exponential,support=3)

x = np.linspace(0, 3, 100)
y = np.array([h4(i) for i in x])

plt.plot(x,y)
```
